```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\raw data.csv", sep=';')
df
```

```python
columns_to_convert = [
    "% response CTG",
    "% response Cell number",
    "% response Nuclear area",
    "% response Nuclear intensity",
    "% response Nuclear membrane permeability"
]

df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.str.replace(',', '.').astype(float))
```

```python
# Definindo um dicionário com os limiares para cada coluna
thresholds = {
    '% response CTG': 30,
    '% response Cell number': 52,
    '% response Nuclear area': 25,
    '% response Nuclear intensity': 11,
    '% response Nuclear membrane permeability': 8
}

# Função para aplicar a transformação com base no limiar
def binarize(column, threshold):
    return column.apply(lambda x: 1 if x >= threshold else 0)

# Aplicando a função para cada coluna especificada
for column, threshold in thresholds.items():
    df[column] = binarize(df[column], threshold)

df
```

```python
# Selecionar e salvar as colunas específicas do DataFrame em arquivos CSV diferentes
df[['Compound ID', 'CXSMILES', 'Vendor', '% response CTG']].to_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\CTG_binary.csv", sep=';', index=False)
df[['Compound ID', 'CXSMILES', 'Vendor', '% response Cell number']].to_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\Cell_number_binary.csv", sep=';', index=False)
df[['Compound ID', 'CXSMILES', 'Vendor', '% response Nuclear area']].to_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\Nuclear_area_binary.csv", sep=';', index=False)
df[['Compound ID', 'CXSMILES', 'Vendor', '% response Nuclear intensity']].to_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\Nuclear_intensity_binary.csv", sep=';', index=False)
df[['Compound ID', 'CXSMILES', 'Vendor', '% response Nuclear membrane permeability']].to_csv(r"D:\OneDrive\Documentos\LabMol\IC-Citotoxicidade\DeepCytosafe multiparam\Data\Nuclear_membrane_permeability_binary.csv", sep=';', index=False)
```
